#### EXCEPTION HANDLING

```python
# try and except
# try -> all risky codes
#except -> return that value
```

```python
ip=int(input("Enter a number"))
def something(ip):
  1/ip
  print('A')

something(ip)
print('B')
```

```python
try:
  something(ip)
except:
  print("B")
```

```python
try:
  something(ip)
except Exception as e:
  print(e)
```

```python
#finally it always run
try:
  print('a')
except ZeroDivisionError as z:
  print(z)
except Exception as e:
  print(e)
finally:
  print("Finally")
```

```python
try:
  print (1/0)
except ZeroDivisionError as z:
  print(z)
except Exception as e:
  print(e)
finally:
  print("Finally")
```

```python
class MyOwnFancyExceptionDoesNothing(Exception):
  def __init__(self,message):
    super().__init__(message)

name="ABCD"

try:
  if len(name)<5:
    raise MyOwnFancyExceptionDoesNothing("Length must be more than 4")
except MyOwnFancyExceptionDoesNothing as e:
  print(e)
```
